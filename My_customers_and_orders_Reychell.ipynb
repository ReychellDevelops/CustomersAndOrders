**Question 1:** How many customers are in the file?
(as help, we have added some comments and starter code to help you structure the solution)

```python
from os.path import exists #importa el modulo exists que verifica si un archivo o directorio existe
import csv

# check if file can be found and open the file
datafile = "./sample_data/customers.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!") #el raise sirve para crear una exepcion, en este caso SystemExit que detiene el programa y muestra un mensaje

# customers file is in the sample_data folder and can be opened.
# Using csv reader is one way of dealing with the file

#how many customers are in the file?

#A list is created where the customer id will be stored without duplication (the set allows this)
uniqueCustomers = set()

#Open the file and allow it to be read.
with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')

  #skip the first row that contains the headers
  next(csvreader)

  #create a for loop to go through each row of the file and count them
  for customer in csvreader:
    #fill the uniqueCustomers list with each customer id (It is located in position 0 of each row)
    customerId = customer[0]
    uniqueCustomers.add(customerId)

fl.close()

#count the values of the uniqueCustomers list which would be the total number of customers
print(f"In total, there are {len(uniqueCustomers)} customers")
print(f"Customers by ID: {uniqueCustomers}")
```

**Question 2:** In how many different states do the customers live in?

```python

#Create a list that stores the states where customers live
uniqueStates = set()
# open file and init vars
with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=",")
  next(csvreader) #skip headers
  #loop to go through each file row
  for state in csvreader:
    states = state[4] #store the values ​​of position 4 of each row in the variable states

    #add to uniqueStates each value store in states
    uniqueStates.add(states.upper().strip()) #Use upper' to make all values ​​uppercase and strip to remove spaces, so that there are no repeated values.
fl.close()

print(f"The customers live in {len(uniqueStates)} kind of states.")
print(f"These states are {uniqueStates}")
```

**Question 3** What is the state with most customers?

```python
# Dictionary to store unique customers by state
statesCustomers = {}

#open file and read data
with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=",")

    #skip headers
    next(csvreader)

    #loop through each row in the CSV file
    for row in csvreader:
        customerId = row[0].strip() #customer id at position 0
        stateCode = row[4].upper().strip() #state code at position 4

        #if state doesn't exist in dictionary, create a new set for that state
        if stateCode not in statesCustomers:
            statesCustomers[stateCode] = set()

        #aadd customer id to the set for the corresponding state
        #the set automatically removes duplicate customer ids
        statesCustomers[stateCode].add(customerId)

# Variables to find the state with the most unique customers
maxState = ""#will store the name of the state with most customers
maxCount = 0 #will store the maximum customer count

#loop through all states and their customer sets
for stateCode, customerSet in statesCustomers.items():
    countByState = len(customerSet)  # Count unique customers in this state

    #if this state has more customers than current maximum, update
    if countByState > maxCount:
        maxCount = countByState
        maxState = stateCode

# Display result
print(f"The state with the most customers is: {maxState} with {maxCount} customers")
print("All done!")
```

**Question 4** What is the state with the least customers?

```python
statesCustomersMin = {}

#open file and init vars
with open(datafile, 'r') as fl:
    csvReader = csv.reader(fl, delimiter=",")

    next(csvReader)#skip headers

    #loop through each row in the file
    for row in csvReader:
        stateCode = row[4].upper().strip() #state position
        customerId = row[0].strip() #id position

        if stateCode not in statesCustomersMin:
            statesCustomersMin[stateCode] = set()

        #add customer ID to the set for the corresponding state
        #the set automatically removes duplicate customer IDs
        statesCustomersMin[stateCode].add(customerId)  #fixed variable name

#variables to find the state with least customers
minStates = []    # will store the name of the state with least customers
minCount = float('inf')  #initialize with infinity to find minimum

#loop through all states and their customer sets
for stateCode, customerSet in statesCustomersMin.items():
    countByState = len(customerSet)  #count unique customers in this state

    #if this state has fewer customers than current minimum, update
    if countByState < minCount:
        minCount = countByState
        minStates = [stateCode]
    elif countByState == minCount:
        minStates.append(stateCode)  # Add to existing minimum states

#display result
print(f"States with the least customers ({minCount} customers): {', '.join(sorted(minStates))}")
```

**Question 5:** What is the most common last name?

```python
lastNames= {}
# open file and init vars
with open(datafile,'r') as fl:
  csvreader = csv.reader(fl, delimiter=",")

  next(csvreader) #skip headers

  for row in csvreader:
    lastName = row[2].upper().strip() #lastname position
    customerId = row[0].strip() #id position

    if lastName not in lastNames:
      lastNames[lastName] = set()

    #add lastname whith its customer id to lastNames list
    lastNames[lastName].add(customerId)

  maxLastN = ""
  maxLCount = 0

  for lastN, customerId in lastNames.items():
    countLast = len(customerId) #Count of customers stored by last name

    """If the customer count for the evaluated last name is greater than the one stored in maxLcount,
    maxLCount is assigned the value of the latter, and maxLastN is assigned the value of the row being evaluated
    in the current cycle"""
    if countLast > maxLCount:
      maxLCount = countLast
      maxLastN = lastN

  #show result
  print(f"The most common last name is {maxLastN} with {maxLCount} customers")
# evaluate the contents of the file
```

**Question #1: How many unique orders are in the orders.csv file?**

```python
#Question #1: How many unique orders are in the orders.csv file?
# check if file can be found and open the file
datafile2 = "./sample_data/orders.csv"
if not exists(datafile2):
  raise SystemExit("You should run the first code cell and download the dataset files!") #el raise sirve para crear una exepcion, en este caso SystemExit que detiene el programa y muestra un mensaje

# open and evaluate orders.csv
with open(datafile2, 'r') as fl2:
  csvreader2 = csv.reader(fl2, delimiter=",")

  next(csvreader2) #skip headers

  #A list is created where the order id will be stored without duplication (the set allows this)
  uniqueOrders = set() #

  for row in csvreader2:
    idOrders = row[1].strip() #id order position with strip for delete spaces

    #if the order id doesnt in the uniqueOrders list, add idOrder to this list
    if idOrders not in uniqueOrders:
      uniqueOrders.add(idOrders)

  print(f"There are {len(uniqueOrders)} unique orders in the file")
```

**Question #2: What is the average number of items per order (rounded to two decimal places)?**

```python
orderItems = {} #dictionary to store items per order

with open(datafile2, 'r') as fl:
  csvreader2 = csv.reader(fl, delimiter=",")

  next(csvreader2)

  for row in csvreader2:
    idOrder = row[1] #id position

    #si la orden no esta dentro de la lista orderItems, se agrega y se inicializa en 0
    if idOrder not in orderItems:
      orderItems[idOrder] = 0 #order id position

    orderItems[idOrder] += 1 #count each product as one item in the order

#average
totalItems = sum(orderItems.values())
totalOrders = len(orderItems)
averageI = (totalItems/totalOrders)

print(f"the average number of items per order is {averageI:.2f}")
```

**Question #3: What is the highest number of items per order?**

```python
itemsOrderH = {}

with open(datafile2, 'r') as fl2:
  csvreader2 = csv.reader(fl2, delimiter=",")

  next(csvreader2) #skip headers

  for row in csvreader2:
    orderId = row[1] #id position

    if orderId not in itemsOrderH:
      itemsOrderH[orderId] = 0

    itemsOrderH[orderId] += 1

  #definimos variables que almacenan cantidad de ordenes mayor y el id de la orden
  highestN = 0
  highestOrder = ''

  #loop para recorrer los items del diccionario
  for order, item in itemsOrderH.items():
    if item > highestN: #si el item evaluado en ese momento es mayor al que esta almacenado, entonces se le asigna ese valor a highestN
      highestN = item
      highestOrder = order

  print(f"the highest items per order is {highestN} from id {highestOrder}")

```

**Question #4: What is the number of orders placed in October 2021?**

```python
octoberOrders = set()

with open(datafile2, 'r') as fl2:
  csvreader2 = csv.reader(fl2, delimiter=",")

  next(csvreader2) #skip headers
  totalOrders = 0

  for row in csvreader2:
    idOrder = row[1] #id position
    dateOrder = row[2] #date position

    #if "2021-10" is in dateOrder
    if "2021-10" in dateOrder:
      octoberOrders.add(dateOrder) #add dateOrder to octoberOrders List

  print(f"The number of order placed in october 2021 are {len(octoberOrders)}")
```

**Question #5: Which customer spent the most amount of money in 2021?**

```python
mostAmount = {}
processedOrders = set()

with open(datafile2, 'r') as fl2:
    csvreader2 = csv.reader(fl2, delimiter=",")
    next(csvreader2)  #skip headers

    for row in csvreader2:
        idCustomer = row[0]  #id customer position
        orderId = row[1]  #id order position
        totalAmount = float(row[3])  # orderTotal position
        dateOrder = row[2]  #dateOrder position

        if "2021" in dateOrder:
            #create unique key for each order
            orderKey = f"{idCustomer}_{orderId}"

            #only process each order once
            if orderKey not in processedOrders:
                #Initialize customer total if not exists
                if idCustomer not in mostAmount:
                    mostAmount[idCustomer] = 0

                #add the order total to customer's spending
                mostAmount[idCustomer] += totalAmount
                processedOrders.add(orderKey)

    #find customer with maximum spending
    maxAmount = 0
    idMax = ''

    for customer, amount in mostAmount.items():
        if amount > maxAmount:
            maxAmount = amount
            idMax = customer

    print(f"The customer that spent the most money in 2021 was ID {idMax} with a total of ${maxAmount:.2f}")
```

**Question #6: Historically, what is the best month for sales?**

```python
monthSales = {}
processedOrders = set()

with open(datafile2, 'r') as fl2:
    csvreader2 = csv.reader(fl2, delimiter=",")
    next(csvreader2)  #skip headers

    for row in csvreader2:
        orderId = row[1]  #id order position
        totalAmount = float(row[3])  #orderTotal position
        dateOrder = row[2]  #dateOrder position

        #only process each order once
        if orderId not in processedOrders:
            #extract month from date (positions 5-7: YYYY-MM-DD)
            month = dateOrder[5:7]

            #initialize month total if not exists
            if month not in monthSales:
                monthSales[month] = 0

            #add the order total to month's sales
            monthSales[month] += totalAmount
            processedOrders.add(orderId)

#find month with maximum sales
maxSales = 0
bestMonth = ''

for month, sales in monthSales.items():
    if sales > maxSales:
        maxSales = sales
        bestMonth = month

#convert month number to name
monthNames = {
    '01': 'January', '02': 'February', '03': 'March', '04': 'April',
    '05': 'May', '06': 'June', '07': 'July', '08': 'August',
    '09': 'September', '10': 'October', '11': 'November', '12': 'December'
}

print(f"Historically, the best month for sales is: {monthNames[bestMonth]}")
print(f"Total Sales: ${maxSales:,.2f}")
```
